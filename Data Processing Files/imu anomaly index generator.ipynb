```python
filename = "P19V1_F5.csv"
channel = "az1"
```

```python
import os
import pandas as pd

# Define the directory path
directory = "..\TFO Data\IMU"

# Get the list of files in the directory
files = os.listdir(directory)


file_path = os.path.join(directory, filename)
    
df = pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt

plt.plot(df[channel])
plt.xlabel('Index')
plt.ylabel(channel)
plt.title('Plot of ' + channel)

plt.xlim(df.index.min(), df.index.max())

plt.show()
```

```python
t1 = 25550
t2 = 25750

plt.plot(df[channel][t1:t2])
plt.xlabel('Index')
plt.ylabel(channel)
plt.title('Zoomed Plot of ' + channel)

plt.xlim(t1, t2)

plt.show()
```
