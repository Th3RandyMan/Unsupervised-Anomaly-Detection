```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Select a random file from the file list
random_file = random.choice(file_list)

# Read the CSV file into a pandas dataframe
df = pd.read_csv(random_file)

# Select a random channel from the dataframe
random_channel = random.choice(df.columns)

Fs = 8000
t = np.linspace(0, len(df[random_channel])/Fs, len(df[random_channel]))

# Plot the selected channel
plt.plot(t, df[random_channel])
plt.title(f"Random Channel from {random_file}")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.xlim(0, t[-1])
plt.show()
```

```python
import matplotlib.pyplot as plt

Fs = 8000

# Plot the spectrogram
plt.specgram(df[random_channel], Fs=Fs)
plt.title('Spectrogram')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.colorbar(label='Intensity')
plt.ylim(0, 400)  # Set the y-axis limit to zoom down to 0 Hz to 400 Hz
plt.show()

fig = plt.gcf()  # Get the current figure
fig_size = fig.get_size_inches()  # Get the size of the figure in inches
print(f"Figure size: {fig_size}")
```
