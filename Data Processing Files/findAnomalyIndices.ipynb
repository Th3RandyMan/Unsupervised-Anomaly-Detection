```python
import glob
import pandas as pd

# Path to the folder containing the clean CSV files
clean_folder_path = "..\TFO Data\IMU\Clean Data"

# Get the list of CSV files in the folder
clean_file_list = glob.glob(clean_folder_path + "\*.csv")
```

```python
# Path to the folder containing the original CSV files
original_folder_path = "..\TFO Data\IMU"

# Get the list of CSV files in the folder
original_file_list = glob.glob(original_folder_path + "\*.csv")
```

```python
import re

file_channel_list = []

for clean_file in clean_file_list:
    # Extract the file name from the clean file path
    file_name = clean_file.split("\\")[-1]

    file_name = re.sub(r"_\d+\.csv$", "", file_name)
    if file_name not in file_channel_list:
        file_channel_list.append(file_name)

print(file_channel_list)
```

```python
entry_num = 3

file_channel_pair = file_channel_list[entry_num]

# Extract the file name from the file_channel_pair
file_name = "_".join(file_channel_pair.split("_")[:-1]) + ".csv"
channel_name = file_channel_pair.split("_")[-1]
```

```python
import matplotlib.pyplot as plt

file_path = "..\\TFO Data\\IMU\\" + file_name

# Read the CSV file into a pandas dataframe
df = pd.read_csv(file_path)

# Plot the column specified by channel_name
plt.plot(df[channel_name])

# Add labels and title to the plot
plt.xlabel('Index')
plt.ylabel(channel_name)
plt.title('Plot of ' + channel_name)

clean_file_name = "P20V1_20_ay1_0.csv"
clean_file_path = "..\\TFO Data\\IMU\\Clean Data\\" + clean_file_name
clean_df = pd.read_csv(clean_file_path)
clean_df_length = len(clean_df)
print(clean_df_length)

# Add a dotted red line at the index of the length
plt.axvline(x=clean_df_length, color='r', linestyle='dotted')

# Display the plot
plt.show()
```
