```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Select 10 random files from the file list
random_files = random.sample(file_list, 10)

Fs = 2000 #Hz

# Iterate over each random file
for file_path in random_files:
    # Read the CSV file into a pandas dataframe
    df = pd.read_csv(file_path)
    
    # Select 10 random channels from the dataframe
    channel = random.sample(list(df.columns), 1)
    
    # Plot the selected channels
    selected_column = df[channel]
    title = file_path.split("\\")[-1]
    print(f"Plot of {title} - {channel}")

    fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 5))

    # Plot the time series
    t = np.arange(0, len(selected_column)/Fs, 1/Fs)
    axes[0].plot(t, selected_column)
    axes[0].set_title(channel)
    axes[0].set_xlabel('Time (s)')

    # Plot the spectrogram
    print(selected_column.shape)
    axes[1].specgram(selected_column, Fs=Fs)
    axes[1].set_xlabel('Time')
    axes[1].set_ylabel('Frequency')
    axes[1].set_title(f'Spectrogram of {channel}')

    plt.tight_layout()
    plt.show()
```
